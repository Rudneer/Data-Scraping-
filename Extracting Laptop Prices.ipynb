```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
pd.set_option('display.max_rows',2000)
```

```python
descriptions=[]
prices=[]
```

```python
for i in range(1,43):
    url="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page="+str(i)
    req=requests.get(url)
    content=BeautifulSoup(req.content,'html.parser')
    desc = content.find_all('div',{'class':'_4rR01T'})
    price = content.find_all('div', {'class':'_30jeq3 _1_WHN1'})
    rating = content.find_all('div', {'class':'_3LWZlK'})
    
    for i in desc:
        descriptions.append(i.text)
    for i in price:
        prices.append(i.text)
```

```python
print(len(descriptions))
print(len(prices))
```

```python
data = {'Description':descriptions,'Price':prices}
df = pd.DataFrame(data)
```
